```python
# 🧾 Encabezado institucional y enunciado del problema
from IPython.display import display, Markdown

info = """
# Universidad Estatal Amazónica
### Facultad de Ciencias Agropecuarias
### Carrera de Ingeniería en Agropecuaria
**Simulación del Crecimiento de una vaca Holstein**
Autor: *Dagni Encalada Sapatanga*

---

> En una finca ganadera del sector **San Juan del Cid**, se está monitoreando el crecimiento del peso de una vaca Holstein durante los primeros meses de vida para evaluar la efectividad del plan nutricional que tiene esta finca.
> Los técnicos han determinado que el peso de la vaca, medido en kilogramos, depende del tiempo en meses desde su nacimiento.
> Al principio el peso aumenta rápidamente, pero conforme la vaca se acerca a la madurez, la tasa de aumento disminuye.
> El peso inicial de la vaca es **40 kg**, con un aumento de **21 kg por mes**, y una reducción progresiva en el crecimiento proporcional al **cuadrado del tiempo transcurrido**, representada por el término **−0.7𝑡²**.

---

**Problemas a resolver:**
a) ¿Cuál es la tasa de cambio del peso de la vaca al mes 6 (o cualquier mes que elijas)?
b) ¿En qué mes crees que la vaca alcanza su peso máximo según este modelo?
c) ¿Cuál sería ese peso máximo?
"""
display(Markdown(info))
# 📦 Librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider
from PIL import Image
from google.colab import files

# 🔽 Subida de imágenes: ternera, joven y adulta
print("📁 Sube las tres imágenes: vaca_ternera.png, vaca_joven.png, vaca_adulta.png")
uploaded = files.upload()

# 🖼️ Carga por nombre
vaca_ternera = Image.open("vaca_ternera.png")
vaca_joven   = Image.open("vaca_joven.png")
vaca_adulta  = Image.open("vaca_adulta.png")

# 📐 Funciones del modelo
def modelo_peso(t, w0, a, b):
    return w0 + a * t - b * t**2

def tasa_cambio(t, a, b):
    return a - 2 * b * t

def seleccionar_imagen(peso):
    if peso <= 60:
        return vaca_ternera
    elif peso <= 90:
        return vaca_joven
    else:
        return vaca_adulta

# 🎨 Visualización integrada
def graficar(w0=40.0, a=21.0, b=0.7, mes_tasa=6):
    t = np.linspace(0, 12, 100)
    peso = modelo_peso(t, w0, a, b)
    peso_mes = modelo_peso(mes_tasa, w0, a, b)
    tasa_mes = tasa_cambio(mes_tasa, a, b)
    t_max = a / (2 * b)
    peso_max = modelo_peso(t_max, w0, a, b)

    # Selección y escalado de imagen
    imagen = seleccionar_imagen(peso_mes)
    escala = peso_mes / 150
    ancho = int(imagen.width * escala)
    alto = int(imagen.height * escala)
    vaca_escalada = imagen.resize((ancho, alto))

    fig, ax = plt.subplots(1, 2, figsize=(16, 6))

    # 📈 Gráfico del modelo
    ax[0].plot(t, peso, color='green', label='Peso (kg)')
    ax[0].axvline(x=t_max, linestyle='--', color='red',
                  label=f'Máximo: {peso_max:.1f} kg en mes {t_max:.1f}')
    ax[0].scatter([mes_tasa], [peso_mes], color='blue', s=100,
                  label=f'Tasa mes {mes_tasa}: {tasa_mes:.1f} kg/mes')
    ax[0].plot([], [], ' ', label=f'Peso inicial: {w0:.1f} kg')
    ax[0].set_title(f'📈 Crecimiento del peso\nPeso inicial: {w0:.1f} kg')
    ax[0].set_xlabel('Meses')
    ax[0].set_ylabel('Peso (kg)')
    ax[0].legend()
    ax[0].grid(True)

    # 🐄 Imagen de vaca
    ax[1].imshow(vaca_escalada)
    ax[1].axis('off')
    ax[1].set_title(f'🐄 Etapa visual: {peso_mes:.1f} kg en mes {mes_tasa}')

    plt.tight_layout()
    plt.show()

# 🎛️ Controles interactivos
interact(graficar,
         w0=FloatSlider(value=40, min=20, max=100, step=1, description='Peso Inicial'),
         a=FloatSlider(value=21, min=10, max=40, step=0.5, description='Crecimiento mensual'),
         b=FloatSlider(value=0.7, min=0.1, max=2.5, step=0.1, description='Reducción cuadrática'),
         mes_tasa=IntSlider(value=6, min=0, max=12, step=1, description='Mes a analizar'))
```
